```python
def caesar_cipher(text, shift):
    return ''.join(chr((ord(char) - 97 + shift) % 26 + 97) if char.islower() else
                   chr((ord(char) - 65 + shift) % 26 + 65) if char.isupper() else char
                   for char in text)

text = "Hello World"
shift = 3
encoded_text = caesar_cipher(text, shift)
print(encoded_text)
```

```python
decoded_text = caesar_cipher(encoded_text, -shift)
print(decoded_text)
```

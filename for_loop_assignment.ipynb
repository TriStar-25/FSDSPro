```python
## 1. Print numbers from 1 to 10 using a `for` loop

for num in range(1, 10):
    print(num)
```

```python
# 2. Calculate the sum of all numbers in a list using a `for` loop

numbers = [1, 2, 3, 4, 5]
sum_of_numbers = 0
for num in numbers:
    sum_of_numbers += num
print("Sum of list elements:", sum_of_numbers)
```

```python
# 3. Print the characters of a string in reverse order using a `for` loop

text = "Hello"
for char in reversed(text):
    print(char, end='')
```

```python
## 4. Find the factorial of a given number using a `for` loop
num = 5
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print(f"Factorial of {num} is {factorial}")
```

```python
## 5. Print the multiplication table of a given number using a `for` loop
number = 3
for i in range(1, 13):
    print(f"{number} x {i} = {number * i}")
```

```python
## 6. Count the number of even and odd numbers in a list using a `for` loop
numbers = [1, 2, 3, 4, 5, 6]
even_count = 0
odd_count = 0
for num in numbers:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
print("Even numbers:", even_count)
print("Odd numbers:", odd_count)
```

```python
## 7. Print the squares of numbers from 1 to 5 using a `for` loop
for num in range(1, 6):
    print(f"Square of {num} is {num ** 2}")
```

```python
## 8. Find the length of a string without using the `len()` function
text = "Hello World"
length = 0
for _ in text:
    length += 1
print("Length of the string:", length)
```

```python
## 9. Calculate the average of a list of numbers using a `for` loop
numbers = [10, 20, 30, 40, 50]
total_sum = 0
for num in numbers:
    total_sum += num
average = total_sum / len(numbers)
print("Average of the list:", average)
```

```python
## 10. Print the first `n` Fibonacci numbers using a `for` loop
n = 10
fib_sequence = [0, 1]
for i in range(2, n):
    next_fib = fib_sequence[i - 1] + fib_sequence[i - 2]
    fib_sequence.append(next_fib)
print(f"First {n} Fibonacci numbers: {fib_sequence}")
```

```python
## 11. Check if a given list contains any duplicates using a `for` loop
numbers = [1, 2, 3, 4, 2]
seen = set()
has_duplicates = False
for num in numbers:
    if num in seen:
        has_duplicates = True
        break
    seen.add(num)
print("Contains duplicates:", has_duplicates)
```

```python
## 12. Print the prime numbers in a given range using a `for` loop
start = 10
end = 50
print("Prime numbers between", start, "and", end, ":")
for num in range(start, end + 1):
    if num > 1:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                break
        else:
            print(num)
```

```python
## 14. Find the maximum element in a 2D list using a nested `for` loop
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
max_element = matrix[0][0]
for row in matrix:
    for element in row:
        if element > max_element:
            max_element = element
print("Maximum element:", max_element)
```

```python
## 17. Convert a list of Fahrenheit temperatures to Celsius using a `for` loop
fahrenheit_temps = [32, 68, 104, 122]
celsius_temps = []
for temp in fahrenheit_temps:
    celsius = (temp - 32) * 5 / 9
    celsius_temps.append(celsius)
print("Temperatures in Celsius:", celsius_temps)
```

```python
## 19. Print the pattern of right-angled triangles using a `for` loop
rows = 5
for i in range(1, rows + 1):
    print('*' * i)
```

```python
## 20. Find the greatest common divisor (GCD) of two numbers using a `for` loop
a = 56
b = 98
gcd = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        gcd = i
print("GCD of", a, "and", b, "is", gcd)
```

```python
## 31. Generate a list of prime numbers using the Sieve of Eratosthenes algorithm and list comprehension
n = 50
primes = [True] * (n + 1)
primes[0] = primes[1] = False
for i in range(2, int(n ** 0.5) + 1):
    if primes[i]:
        for j in range(i * i, n + 1, i):
            primes[j] = False
prime_numbers = [i for i in range(n + 1) if primes[i]]
print("Prime numbers up to", n, ":", prime_numbers)
```

```python
## 32. Generate a list of all Pythagorean triplets up to a specified limit using list comprehension
limit = 20
pythagorean_triplets = [(a, b, c) for a in range(1, limit) for b in range(a, limit) for c in range(b, limit) if a ** 2 + b ** 2 == c ** 2]
print("Pythagorean triplets up to", limit, ":", pythagorean_triplets)
```

```python
## 40. Generate a list of prime palindromic numbers using list comprehension
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def is_palindrome(num):
    return str(num) == str(num)[::-1]

limit = 1000
prime_palindromic_numbers = [num for num in range(2, limit + 1) if is_prime(num) and is_palindrome(num)]
print("Prime palindromic numbers up to", limit, ":", prime_palindromic_numbers)
```

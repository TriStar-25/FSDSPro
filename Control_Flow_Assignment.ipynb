```python
# 20. Determine the roots of a cubic equation using Cardano's formula
import cmath

def cubic_roots(a, b, c, d):
    if a == 0:
        return "Not a cubic equation"

    # Normalize coefficients
    b /= a
    c /= a
    d /= a

    # Calculate intermediary values
    delta_0 = b**2 - 3 * c
    delta_1 = 2 * b**3 - 9 * b * c + 27 * d

    C = cmath.sqrt((delta_1**2 - 4 * delta_0**3)) / 2

    # Calculate the three cube roots
    omega = cmath.exp(2j * cmath.pi / 3)
    roots = []
    for k in range(3):
        root = -1 / 3 * (b + omega**k * C)
        roots.append(root)

    return roots

print(cubic_roots(1, -6, 11, -6))  # Test example
```

```python
# 34. Matrix multiplication using nested loops
def matrix_multiply(matrix_a, matrix_b):
    rows_a, cols_a = len(matrix_a), len(matrix_a[0])
    rows_b, cols_b = len(matrix_b), len(matrix_b[0])

    if cols_a != rows_b:
        return "Matrix multiplication not possible!"

    result = [[0] * cols_b for _ in range(rows_a)]

    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]
    return result

matrix_a = [[1, 2], [3, 4]]
matrix_b = [[2, 0], [1, 3]]
print(matrix_multiply(matrix_a, matrix_b))
```

```python
# 35. Basic text-based Tic-Tac-Toe game against the computer
import random

def display_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 5)

def is_winner(board, mark):
    return any(all(cell == mark for cell in row) for row in board) or \
           any(all(row[col] == mark for row in board) for col in range(3)) or \
           all(board[i][i] == mark for i in range(3)) or \
           all(board[i][2 - i] == mark for i in range(3))

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    moves = [(i, j) for i in range(3) for j in range(3)]
    player_mark, computer_mark = "X", "O"

    while moves:
        display_board(board)
        row, col = map(int, input("Enter row and column (0-2): ").split())
        if (row, col) not in moves:
            print("Invalid move, try again.")
            continue

        board[row][col] = player_mark
        moves.remove((row, col))
        if is_winner(board, player_mark):
            display_board(board)
            print("You win!")
            break

        if not moves:
            print("It's a tie!")
            break

        computer_move = random.choice(moves)
        board[computer_move[0]][computer_move[1]] = computer_mark
        moves.remove(computer_move)

        if is_winner(board, computer_mark):
            display_board(board)
            print("Computer wins!")
            break

tic_tac_toe()
```

```python
# 36. Fibonacci numbers using iterative methods
def fibonacci_iterative(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=' ')
        a, b = b, a + b
    print()

fibonacci_iterative(10)
```

```python
# 37. Fibonacci using memoization
def fibonacci_memo(n, memo={}):
    if n in memo:
        return memo[n]
    if n <= 1:
        return n
    memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]

print(fibonacci_memo(10))  # Test example
```

```python
# 39. Basic text-based blackjack game against the computer
import random

def blackjack():
    def draw_card():
        return random.randint(1, 11)

    player_score = draw_card() + draw_card()
    computer_score = draw_card() + draw_card()
    print(f"Player's initial score: {player_score}")

    while player_score < 21:
        action = input("Hit or Stay? ").lower()
        if action == "hit":
            player_score += draw_card()
            print(f"Player's score: {player_score}")
        else:
            break

    print(f"Computer's score: {computer_score}")

    if player_score > 21:
        print("Bust! You lose.")
    elif computer_score > 21 or player_score > computer_score:
        print("You win!")
    else:
        print("You lose!")

blackjack()
```

```python
# 40. Prime factors using trial division
def prime_factors(n):
    factors = []
    divisor = 2
    while n > 1:
        while n % divisor == 0:
            factors.append(divisor)
            n //= divisor
        divisor += 1
    return factors

print(prime_factors(56))  # Test example
```
